```python
def missing_values(new_data):
    total=new_data.isnull().sum().sort_values(ascending=False)
    percentage=(new_data.isnull().sum()/new_data.isnull().count()).sort_values(ascending=False)
    percentage_final=(round(percentage,2)*100)
    total_percentage=pd.concat(objs=[total,percentage_final],axis=1,keys=['total','%'])
    return total_percentage
```

# skewness check 

```python
from scipy.stats import skew
for i in new_data:
    print(skew(new_data[i]))
```

```python
from sklearn.preprocessing import power_transform
def skewedness(data):
    skewed=[]
    for col in data:
        sk=skew(data[col])
        if(sk>1.9 or sk<=-1.9):
            skewed.append(col)
    return skewed
skewd_values=skewedness(new_data)
skewd_values
```

```python
inp_arry=new_data
```

```python
from scipy import stats

Total_Bilirubin = stats.boxcox(inp_arry['Total_Bilirubin'])[0]
Direct_Bilirubin = stats.boxcox(inp_arry['Direct_Bilirubin'])[0]
Alkaline_Phosphotase = stats.boxcox(inp_arry['Alkaline_Phosphotase'])[0]
Alamine_Aminotransferase = stats.boxcox(inp_arry['Alamine_Aminotransferase'])[0]
Aspartate_Aminotransferase = stats.boxcox(inp_arry['Aspartate_Aminotransferase'])[0]

dataset = pd.DataFrame({'Total_Bilirubin': Total_Bilirubin, 'Direct_Bilirubin': Direct_Bilirubin,'Alkaline_Phosphotase':Alkaline_Phosphotase,
                       'Alamine_Aminotransferase':Alamine_Aminotransferase,'Aspartate_Aminotransferase':Aspartate_Aminotransferase})

dataset[['Age', 'Gender','Total_Protiens', 'Albumin','Albumin_and_Globulin_Ratio', 'target']] = inp_arry[['Age', 'Gender','Total_Protiens', 'Albumin','Albumin_and_Globulin_Ratio', 'target']]

new_df = pd.DataFrame()

new_df[['Age', 'Gender', 'Total_Bilirubin', 'Direct_Bilirubin',
       'Alkaline_Phosphotase', 'Alamine_Aminotransferase',
       'Aspartate_Aminotransferase', 'Total_Protiens', 'Albumin',
       'Albumin_and_Globulin_Ratio', 'target']] = dataset[['Age', 'Gender', 'Total_Bilirubin', 'Direct_Bilirubin',
       'Alkaline_Phosphotase', 'Alamine_Aminotransferase',
       'Aspartate_Aminotransferase', 'Total_Protiens', 'Albumin',
       'Albumin_and_Globulin_Ratio', 'target']]

new_df
```

```python
df_data = new_df
df_data.head()
```

```python
x=df_data.iloc[:,:-1]
```

```python
y=df_data.iloc[:,-1]
```

# feature scaling 

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x,y)
```

```python
from imblearn.over_sampling import SMOTE 
sm = SMOTE(random_state = 30) 
x_res, y_res = sm.fit_sample(x,y) 
```

```python
print(x_res.shape,y_res.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_res,y_res,test_size=0.32,random_state=30)
```

# Logistic Regression

```python
from sklearn.metrics import accuracy_score,f1_score,r2_score,classification_report,confusion_matrix,roc_auc_score
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
print(accuracy_score(y_test,y_predict))
print(f1_score(y_test,y_predict))
print(classification_report(y_test,y_predict))
print(roc_auc_score(y_test,y_predict))
```
